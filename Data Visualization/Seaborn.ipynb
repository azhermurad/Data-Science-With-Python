## The Data

We will be working with a famous titanic data set for these exercises.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.set_style('whitegrid')
sns.set_context(context="paper")
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(data=titanic,x="fare",y="age",hue="who")
```

```python
# Visualization: Survival based on passenger class
sns.countplot(data=titanic,x="survived",hue="pclass")
```

```python

sns.histplot(data=titanic,x="fare",bins=20)
```

```python
# Visualization: Avarage age of passenger based on passenger class
sns.boxplot(data=titanic,x="class",y="age")
```

```python
plt.figure(figsize=(7,7))
sns.swarmplot(data=titanic,x="class",y="age")
```

```python
sns.countplot(data=titanic,x="sex")
```

```python

data =titanic.corr()
sns.heatmap(data)
plt.title("titanic.corr()")
plt.show()
```

```python

sns.displot(data=titanic,x="age", col="sex",row="alive")
```

```python
a =sns.FacetGrid(data=titanic,col="sex")
a.map(sns.histplot,"age")
```

```python
sns.countplot(data=titanic,x="adult_male",hue="alive")
```

```python
# Visualization: Age distribution of passengers
sns.histplot(x='age', data=titanic, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()
```
